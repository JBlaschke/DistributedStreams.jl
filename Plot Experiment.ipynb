```julia
data = sort(collect(log_2["worker_log"]), by = x->x[1])
x = [d[1] for d in data]./60
y = map(x->x == 1 ? 0 : x, [d[2] for d in data])
plot(
    x.-x[1], y,
    linewidth=2, box=:on, tickfont=font(10), guidefont=font(12),
    legend = false
)
```
